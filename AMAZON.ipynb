```python
import os
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, regexp_extract, round

# 1. Iniciar la sesión de Spark
spark = SparkSession.builder.appName("Amazon_Reviews_Final").getOrCreate()

# 2. Definir la ruta a tu carpeta archive (8)
base_path = r"C:\Users\llama\Downloads\archive (8)"
archivos = [f for f in os.listdir(base_path) if f.endswith('.csv')]

if not archivos:
    print("No se encontró el archivo CSV en la carpeta.")
else:
    ruta_completa = os.path.join(base_path, archivos[0])
    print(f" Cargando: {ruta_completa}")

    # 3. EXTRACCIÓN
    df_raw = spark.read.csv(ruta_completa, header=True, inferSchema=True)

    # 4. TRANSFORMACIÓN (Limpieza del error de CAST)
    # Buscamos cualquier dígito (\d) en la columna 'Rating' y lo volvemos Entero
    df_limpio = df_raw.withColumn("Rating_Num", regexp_extract(col("Rating"), r"(\d)", 1).cast("int"))

    # 5. CREACIÓN DE LA TABLA DE HECHOS (Punto 5)
    # Filtramos nulos para evitar que filas malformadas rompan las consultas
    fact_amazon = df_limpio.select(
        col("Reviewer Name").alias("Usuario"),
        col("Country").alias("Pais"),
        col("Rating_Num").alias("Rating"),
        col("Review Date").alias("Fecha")
    ).filter(col("Rating").isNotNull())

    # Registrar vista temporal para SQL
    fact_amazon.createOrReplaceTempView("hechos_amazon")
    print("🚀 ETL Completado con éxito.")
    fact_amazon.show(5)
```

```python
import os
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, regexp_extract, trim

# 1. Iniciar sesión
spark = SparkSession.builder.appName("Amazon_Final_Correction").getOrCreate()

# 2. Ruta
base_path = r"C:\Users\llama\Downloads\archive (8)"
archivo = os.path.join(base_path, "Amazon_Reviews.csv")

# 3. Leer CSV
df_raw = spark.read.csv(archivo, header=True, inferSchema=True)

# 4. Limpiar nombres de columnas (quitar espacios invisibles)
for col_name in df_raw.columns:
    df_raw = df_raw.withColumnRenamed(col_name, col_name.strip())

# 5. TRANSFORMACIÓN SEGURA (Uso de expr y try_cast)
# Primero extraemos el número, y si no puede convertirlo, devuelve NULL en vez de error
df_limpio = df_raw.withColumn("Rating_Num", 
    col("Rating").cast("string") # Aseguramos que sea string para regex
)
df_limpio = df_limpio.withColumn("Rating_Num", 
    regexp_extract(col("Rating_Num"), r"(\d)", 1)
)

# ESTA ES LA LÍNEA CLAVE: Usamos 'expr' para usar try_cast de SQL
from pyspark.sql.functions import expr
df_limpio = df_limpio.withColumn("Rating_Num", expr("try_cast(Rating_Num as int)"))

# 6. CREACIÓN DE TABLA DE HECHOS (Filtramos los que quedaron como NULL)
fact_amazon = df_limpio.select(
    col("Reviewer Name").alias("Usuario"),
    col("Country").alias("Pais"),
    col("Rating_Num").alias("Rating"),
    col("Review Date").alias("Fecha")
).filter(col("Rating").isNotNull()) 

# 7. REGISTRO Y CONSULTAS
fact_amazon.createOrReplaceTempView("hechos_final")

try:
    total = fact_amazon.count()
    print(f"--- REVISIÓN DE DATOS ---")
    print(f"Total de registros válidos cargados: {total}")
    fact_amazon.show(5)

    if total > 0:
        print("\n--- RESULTADO DE LAS CONSULTAS ---")
        # Consulta 1: Promedio
        spark.sql("SELECT 'Promedio Global' as Metrica, ROUND(AVG(Rating), 2) as Valor FROM hechos_final").show()
        # Consulta 2: Países
        spark.sql("SELECT Pais, COUNT(*) as Total FROM hechos_final GROUP BY Pais ORDER BY Total DESC LIMIT 5").show()
        # Consulta 3: Estrellas
        spark.sql("SELECT Rating, COUNT(*) as Cantidad FROM hechos_final GROUP BY Rating ORDER BY Rating DESC").show()
except Exception as e:
    print(f"Algo salió mal: {e}")
```
